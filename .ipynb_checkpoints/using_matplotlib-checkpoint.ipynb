## Scientific Computation

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
from jupyterthemes import jtplot
jtplot.style()

a = np.array(range(100))

plt.figure(0)
plt.plot(a)

plt.figure(1)
plt.plot(a**2,color="green")

plt.figure(3)
plt.plot(a**2,color="green")
plt.plot(a**3,color="red")

plt.show()
```
